## Step 2: Enter Your Settings

Edit the values below to match what you want to crawl, then **run the cell**.

````python
# --- Your URLs (comma-separated) ---
# urls = "https://www.starhub.com/personal.html"
# urls = "https://www.starhub.com/personal.html, https://www.starhub.com/personal/support/article.html?id=UEn42B7Ysx21aZQc7KOWJ7, https://consumer.starhub.com/personal/store/mobile/devices/samsung/galaxy-s26-ultra-5g, https://consumer.starhub.com/personal/store/mobile-plans, https://consumer.starhub.com/personal/store/mobile/devices, https://consumer.starhub.com/personal/store/mobile/devices/apple, https://consumer.starhub.com/personal/store/mobile/devices/samsung, https://consumer.starhub.com/personal/store/mobile/devices/oppo, https://consumer.starhub.com/personal/store/mobile/tablets-watches"
urls = "https://www.ato.gov.au/"

# --- Crawler options ---
exclude_paths = "ato.gov.au/api/"        # URLs matching these patterns will be skipped (leave empty for none)
include_only_paths = "ato.gov.au"   # Only crawl URLs matching these patterns (leave empty for all)
limit = 9999            # Maximum number of pages to crawl
max_depth = 5             # How many clicks deep to follow links
flush_interval = 1       # Save urls.txt to disk every N pages (protects against crashes)
delay = 3                 # Seconds to wait between page crawls — paces the crawl to avoid blocks (0 = no delay)
max_retries = 5           # Retry rounds for blocked pages (0 = no retries)

# --- Page options ---
exclude_tags = "nav, script, form, style"   # HTML elements to remove
include_only_tags = ""                       # Only keep content inside these elements (leave empty for all)
wait_for = 3                                 # Extra seconds after page loads before extracting — gives slow JS time to render (0 or None = skip)
timeout = 60                                 # Hard limit per page — caps the page load + wait_for phase (seconds)
max_file_size_mb = 10                        # Max size of each output file (MB)
extract_main_content = True                  # True = article/product text only (strips menus, footers, sidebars), False = full page
output_extension = ".md"                    # File format: ".txt" or ".md"

# --- Display options ---
activity_log_size = 10                       # Number of recent actions to show in progress bar

print("Settings saved!")
````

## Step 3: Start Crawling

Run the cell below to start crawling. Content is **extracted and saved to disk automatically** as pages are crawled — if the process is interrupted, your progress is safe.

````python
from crawl4md import ContentExtractor, CrawlerConfig, FileWriter, PageConfig, SiteCrawler

crawler_config = CrawlerConfig(
    urls=urls,
    exclude_paths=exclude_paths,
    include_only_paths=include_only_paths,
    limit=limit,
    max_depth=max_depth,
    flush_interval=flush_interval,
    delay=delay,
    max_retries=max_retries,
)
page_config = PageConfig(
    exclude_tags=exclude_tags,
    include_only_tags=include_only_tags,
    wait_for=wait_for,
    timeout=timeout,
    max_file_size_mb=max_file_size_mb,
    extract_main_content=extract_main_content,
    output_extension=output_extension,
)

extractor = ContentExtractor(page_config)
writer = FileWriter(max_file_size_mb=page_config.max_file_size_mb, file_extension=page_config.output_extension)

crawler = SiteCrawler(crawler_config, page_config, output_base="results", extractor=extractor, writer=writer, activity_log_size=activity_log_size)
results = crawler.crawl()
````

## Step 4: View Results

Run the cell below to see a summary of the saved files.

````python
crawler.print_summary(results)
````
